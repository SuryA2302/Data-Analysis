```python
import pandas as pd
import os
```

```python
files = [file for file in os.listdir(r"Z:\DataScience_assets\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data")]

data = pd.DataFrame()

for file in files:
    data1 = pd.read_csv(r"Z:\DataScience_assets\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data\\" + file)
    data = pd.concat([data,data1])
    
```

```python
data.head()
```

#   Data Cleaning

```python
#drop NaN values
data.dropna(inplace= True)
data.tail(100)
```

```python
data.to_csv("Sales_Data.csv",index = False)
```

```python
data["Month"] = data['Order Date'].str[0:2]
data.head()
```

```python
temp = data[data["Month"]=="Or"]  #Checking rows on the basis of error 'Or'
temp
```

```python
data = data[data["Month"]!="Or"]
data.head()
```

```python
data.info()
```

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])
data['Price Each'] = pd.to_numeric(data['Price Each'])
```

```python
data.info()
```

```python
data["Sales"] = data["Quantity Ordered"] * data["Price Each"]
```

```python
data.head()
```

# Q1. Best Month for Sales(in terms of earning)

```python
sales = data.groupby("Month").sum()  #accumulating sales of all 12 months
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

figure = plt.figure(figsize=(5,5))

plt.bar(months,sales['Sales'])
plt.xticks(months)
plt.ylabel("Sales in USD ($)")
plt.xlabel("Month")
plt.show()
```
